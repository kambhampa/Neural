```python


from sklearn import metrics
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

import warnings
from sklearn.exceptions import ConvergenceWarning, UndefinedMetricWarning
```

```python
warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)
```

```python
df=pd.read_csv('C:/Users/saibh/Downloads/glass.csv')
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Type', axis=1), df['Type'], test_size=0.3, random_state=42)

# Create a Linear SVM model and fit it to the training data
svm = LinearSVC(random_state=42)
svm.fit(X_train, y_train)

# Evaluate the model on the test data
score = svm.score(X_test, y_test)
y_pred = svm.predict(X_test)
report = classification_report(y_test, y_pred)

print(f"Accuracy score: ", score)
print(f"Classification report:\n", report)
```
